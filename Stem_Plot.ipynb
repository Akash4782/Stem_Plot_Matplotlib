```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,linefmt = ":")
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,linefmt = "--")
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,linefmt = "-.")
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,linefmt = ":" , markerfmt = "r+")
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,linefmt = ":" , bottom = 1)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,linefmt = ":" , bottom = 9 , markerfmt = "ro")
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,linefmt = ":" , basefmt = "g")
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,linefmt = ":" , basefmt = "g" , label = "Pyton")
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,linefmt = ":" ,markerfmt = "ro" ,bottom = 0, basefmt = "g" , label = "Pyton" )
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]
plt.stem(x,y,linefmt = ":" ,markerfmt = "ro" ,bottom = 0, basefmt = "g" , label = "Pyton" , orientation = "horizontal")
plt.legend()
plt.show()
```
